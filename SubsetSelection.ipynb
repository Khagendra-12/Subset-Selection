# **Subset Selection Program**
**Objective** -->

*   Take user input for a set of numbers.
*   Form subsets from the given set. Subset size is defined by the user.
*   Take user input for the sum of elements of the subset.
*   Take user input for the number of iterations to run.
*   Output with the subsets and their total number under the number of deffined iterations.






**Required Import**

-->  Import combinations from itertools.

```python
from itertools import combinations
```

**Input the elements of the set**

-->  Take user input for the elements of set.

-->  Keep prompting the user if input is invalid.

-->  Turns the input from the user to a list of integer.

```python
while True:
    A = input("Enter the set of numbers(sepreate each number by space): ")
    try:
        A = list(map(int, A.split()))
        break
    except ValueError:
        print("Invalid input. Please enter integers separated by spaces.")
```

**Remove duplicates**

--> Ask the user if its required to remove the duplicates.

--> If removal of duplicates is required turn the list to set and then back to list.

--> Prints of the list of numbers input by the user in the form of a list.

```python
while True:
    q = input("Do you want to remove duplicates(Y/N): ")
    if q.upper() == 'Y':
        A=set(A)
        A=list(A)
        print(f"Set of numbers(After removal of duplicates): {A}")
        break
    elif q.upper() == 'N':
        print(f"Set of numbers: {A}")
        break
    else:
        print("Invalid input.")
        continue
```

**Subset size**

--> Prompt the user to input the required size of the subset.

--> Check to see if valid subset size is entered and re-prompt if the size is incorrect.

```python
while True:
    z = input("Enter subset size range (e.g., '3-5' or '4'): ")
    if '-' in z:
        parts = z.split('-')
        if len(parts) == 2 and all(p.strip().isdigit() for p in parts):
            x, y = sorted(map(int, parts))
            break
        else:
            print("Invalid input. Enter a number or a range like '3-5' or '4'.")
    elif z.strip().isdigit():
        x = y = int(z)
        break
    else:
        print("Invalid input. Enter a number or a range like '3-5' or '4'.")
```

**Sum of elements**

--> Prompt the user to ask the required sum of subset.

```python
while True:
    s = (input("Enter the requireed sum of the subset values: "))
    if s.strip().isdigit():
        s=int(s)
        break
    else:
        print("Invalid input.")
        continue
```

**Number of iterations**

--> Prompt the user to ask the number of iterations to run.

```python
while True:
    iter=(input("Enter the number of iterations to run: "))
    if iter.strip().isdigit():
        iter=int(iter)
        break
    else:
        print("Invalid input.")
        continue
```

**All valid subset**

--> Create a new empty list to hold all the valid subset (subset whos elemnts add up to the given sum).

```python
allvsubset = []
```

**Main condition**

--> Finds all subset in the given range from the list L.

--> If the sum of all elements of subset is equal to given sum add them to allvsubset.

--> Runs the condition till the length of all valid subset is less than or equal to the number of iterations.

```python
for i in range(x, y + 1):
    for subset in combinations(A, i):
        if sum(subset) == s:
            allvsubset.append(subset)
            if len(allvsubset) >= iter-1:
                break
```

**Printing all subset**

--> Prints the number of all valid subset.

--> Prints all valid subsest.

```python
print(f"Total valid subsets: {len(allvsubset)}")
print("[ ", end="")
for subset in allvsubset:
    print(subset, end = ", ")
print("\b\b]")
```
